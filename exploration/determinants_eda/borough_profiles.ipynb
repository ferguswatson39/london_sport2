```python
def query_col(col_name : str, meta : pyreadstat.metadata_container) -> str:
    names = meta.column_names
    idx = names.index(col_name)
    label = meta.variable_to_label[col_name]
    return meta.column_labels[idx], meta.value_labels[label]

y = query_col('inclus_a', meta)
y
```

```python
features = [
    'lifesat', 'lone', 'worthw', 'indev', 'happy', 'comm1', 'comm2',
    'anxious', 'motivex2a', 'motivex2b', 'motivex2c', 'motivex2d', 'inclus_a',
    'inclus_b', 'inclus_c', 'VolAny'
]
features_dict = {
    'lifesat' : 'Life Satisfaction, 0 = Not, 10 = Completely',
    'lone' : 'Feel lonely, 1=Often, 5=Never',
    'worthw' : 'Life worthwhile, 0=Not, 10=Completely',
    'indev' : 'I can achieve goals, 1=SD, 5=SA',
    'happy' : 'Happy yesterday, 0=No, 10=Completely',
    'comm1' : 'Local area trust, 1=SD, 5=SA',
    'comm2' : 'Diverse backgrounds, 1=Disagree, 4=Agree ',
    'anxious' : 'How Anxious yesterday, 0=No, 10=Completely',
    'motivex2a' : 'Excersize for fit/healthy, 1=SA, 5=SD',
    'motivex2b' : 'Excersize for relax/less worry, 1=SA, 5=SD',
    'motivex2c' : 'Excersize socially, fun with friends, 1=SA, 5=SD',
    'motivex2d' : 'Excersize for challenge, 1=SA, 5=SD',
    'inclus_a' : 'Excersize env inclusive, 1=SA, 5=SD',
    'inclus_b' : 'See similar people when excersize, 1=SA, 5=SD', 
    'inclus_c' : 'Feel safe in pub excer places, 1=SA, 5=SD',
    'VolAny' : 'Any vol in last 12, 0=No, 1=Yes'
}

def make_dfs(features):
    dfs = []
    df, meta = pyreadstat.read_sav(path, usecols = the_cols)
    df = df[df['LondInOut'].notna()]
    print(f'Initial DataFrame has length: {len(df)}')
    for f in features:
        series = df.groupby(['LA_2023', f])['MEMS7_ALL'].mean()
        series = series.to_frame()
        series = series.reset_index()
        print(f'Series: {f} has length: {len(series)}')
        dfs.append(series)
    combined = pd.concat(dfs)
    return combined
combined = make_dfs(features)
```
