```python
path = r'C:/Masters/London Sport/activeplacesdatasites_json.json'
places = pd.read_json(path)
```

```python
all = []

df = pd.DataFrame(columns = [
            'ID',
            'name' ,
            'lad_code' ,
            'lad_name' ,
            'facility_id' ,
            'facility_type_id' ,
            'facility_type' ,
            'facility_subtype_id' ,
            'facility_subtype_name' ,
            'year_built' ,
            'year_built_estimated' ,
            'is_refurbished' ,
            'year_refurbished' ,
            'status',
            'closure_date',
            'meets_active_places_criteria' ,
            'associated_facilities'
       
       ])
for i in places['items']:
    for j in i['data']['facilities']:
        new = [
            i['data']['id'],
            i['data']['name'],
            i['data']['localAuthorityCode'],
            i['data']['localAuthorityName'],
            j['id'],
            j['facilityType']['id'],
            j['facilityType']['name'],
            j['facilitySubType']['id'],
            j['facilitySubType']['name'],
            j['yearBuilt'],
            j['yearBuiltEstimated'],
            j['isRefurbished'],
            j['yearRefurbished'],
            j['status']['name'],
            j['closureDate'],
            j['meetsActivePlacesCriteria'],
            len(i['data']['facilities'])
        ]
        
        all.append(new)
        
    
all = pd.DataFrame(all, columns=df.columns)
combined = pd.concat([df, all])
combined.head()
```

```python
combined.to_csv('active_places.csv', index=False)
```
