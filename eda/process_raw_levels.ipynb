```python
########################################
####  PROCESS DATA | RAW TO PICKLE  ####
########################################
import re
from collections import defaultdict
import pyreadstat
import pandas as pd
import pickle
from tqdm.auto import tqdm
```

```python
#### Columns List + Spss Files ####
cols = [
    'MEMS7GR_ALL',
    'wt_final',
    'wt_time',
    'xStrata',
    'Reg9',
    'LondInOut',
    'LA_2023',
    'IMD10',
    'happy',
    'lifesat',
    'worthw',
    'anxious',
    'Motiva_POP',
    'motivb_POP',
    'motivc_POP',
    'motivd_POP',
    'motive_POP',
    'motivex2a',
    'motivex2b',
    'motivex2c',
    'motivex2d',
    'VolAny',
    'VolFrqB_Pop',
    'VolCnt',
    'volint1',
    'volint2',
    'volint3',
    'volint4',
    'volint5',
    'volint6',
    'volint7',
    'WHOWITHA_SPORTCOUNT_A01',
    'WHOWITHB_SPORTCOUNT_A01',
    'WHOWITHC_SPORTCOUNT_A01',
    'WHOWITHD_SPORTCOUNT_A01',
    'Disab3',
    'Gend3',
    'Age9',
    'Eth7',
    'DURATION_SPORTCOUNT_A01',
    'NSSEC5',
    'Educ6',
    'Relig7',
    'Orient4',
    'MEMS7_IN_SPORTCOUNT_A01',
    'MEMS7_OUT_SPORTCOUNT_A01',
]

files = {
    2015: '../../raw data/2015-2016/spss/spss28/active_lives_survey_nov_15-16_data_year_1_shared_20250106.sav',
    2016: '../../raw data/2016-2017/spss/spss28/active_lives_survey_nov_16-17_data_year_2_shared_20250106.sav',
    2017: '../../raw data/2017-2018/spss/spss28/active_lives_survey_nov_17-18_data_year_3_shared_20250103.sav',
    2018: '../../raw data/2018-2019/spss/spss28/active_lives_survey_nov_18-19_data_year_4_shared_20250103.sav',
    2019: '../../raw data/2019-2020/spss/spss28/active_lives_survey_nov_19-20_data_year_5_shared_20250103.sav',
    2020: '../../raw data/2020-2021/spss/spss28/active_lives_survey_nov_20-21_data_year_6_shared_20250103.sav',
    2021: '../../raw data/2021-2022/spss/spss28/active_lives_survey_nov_21-22_data_year_7_shared_20250103.sav',
    2022: '../../raw data/2022-2023/spss/spss28/active_lives_survey_nov_22-23_data_year_8_shared_20250103.sav',
}
```

```python
#### Read File ####
year_dfs = {}
for year in tqdm(files.keys()):
    print(f'Processing Year {year}/{year + 1}')
    file_path = files[year]
    df, meta = pyreadstat.read_sav(file_path)
    london = df[df['LondInOut'].notna()]
    london = london[london['DURATION_SPORTCOUNT_A01'] >= 150]
    available_cols = [c for c in cols if c in london.columns]
    
    level_dfs = {}
    level_cols = defaultdict(list)
    print('Processing Columns')
    for col in tqdm(london.columns):
        # Check if col matches MONTHS_12_* pattern and extract the level letter.
        match = re.search(r'^MONTHS_12_.*_([A-Z])(\d+)$', col)
        if match:
            letter = match.group(1)
            level_cols[letter].append(col)
    
    print('Combining Levels')
    for level in tqdm(level_cols):
        # Select fixed cols + MONTHS_12 cols for this level.
        level_dfs[level] = london[available_cols + level_cols[level]]
    
    year_dfs[year] = level_dfs
    
with open('years_levels_data.pkl', 'wb') as f:
    pickle.dump(year_dfs, f)
```
